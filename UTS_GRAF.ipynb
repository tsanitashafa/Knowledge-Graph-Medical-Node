```python
# Clean the 'penyakitLabel' and 'label' columns by stripping spaces and converting to lowercase
df_wiki['qid'] = df_wiki['qid'].str.strip().str.lower()
df_dbp['qid'] = df_dbp['qid'].str.strip().str.lower()

# Merge the dataframes based on cleaned columns
merged_df = pd.merge(df_wiki, df_dbp, left_on='qid', right_on='qid', how='left')

# Show the merged result
merged_df
```

```python
import pandas as pd

# Assuming df_wiki and df_dbp are already loaded DataFrames

# Perform an inner join on the 'qid' column to get rows with the same qid in both dataframes
df_matching_qid = pd.merge(df_wiki, df_dbp, on='qid', how='left')

# Show the resulting DataFrame with matching qid
df_matching_qid
```

```python
# Melakukan merge antara df_wiki dan df_dbp berdasarkan qid
df_merged = pd.merge(df_wiki, df_dbp, on='qid', how='left', suffixes=('_wiki', '_dbp'))

# Menampilkan hasil merge
df_merged
```

```python
# Gabungkan kolom yang berakhiran _x dan _y menjadi satu kolom saja
df_matching_qid['penyakitLabel'] = df_matching_qid['penyakitLabel_x'].combine_first(df_matching_qid['penyakitLabel_y'])
df_matching_qid['gejalaLabel'] = df_matching_qid['gejalaLabel_x'].combine_first(df_matching_qid['gejalaLabel_y'])
df_matching_qid['kategoriLabel'] = df_matching_qid['kategoriLabel_x'].combine_first(df_matching_qid['kategoriLabel_y'])
df_matching_qid['treatmentLabel'] = df_matching_qid['treatmentLabel_x'].combine_first(df_matching_qid['treatmentLabel_y'])

# Menghapus kolom _x dan _y yang sudah digabung
df_matching_qid.drop(['penyakitLabel_x', 'penyakitLabel_y', 'gejalaLabel_x', 'gejalaLabel_y',
                        'kategoriLabel_x', 'kategoriLabel_y', 'treatmentLabel_x',
                       'treatmentLabel_y'], axis=1, inplace=True)

# Menampilkan hasil setelah penggabungan kolom
df_matching_qid
```
